```python
# id    name            age  subscribed
# 276   John Doe        26   True
# 372   Jane Doe        19   False
# 563   Carrie Dubose   32   False

# dict of lists
user_data_compact = {
# metadata
'columns': ['id', 'name', 'age', 'usage', 'subscribed'],
'length': 3,

# data (column first representation)
'id': [276, 372, 563], # column id
'name': ['John Doe', 'Jane Doe', 'Carrie Dubose'], # column name
'age': [26, 19, 32], # list of ints
'discovery': ['facebook', 'facebook', 'youtube'],
'usage': [187, 37, 257],
'subscribed': [True, False, False],
}

# traversing the table
for row_index in range(user_data_compact['length']):
  for column in user_data_compact['columns']:
    print(f'{column} = {user_data_compact[column][row_index]}')
  print()
```
